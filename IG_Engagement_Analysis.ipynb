### USANDO O GROUPBY PARA ENTENDER MELHOR OS DADOS

```python
#Mudando a formatação para que os dados numéricos contenham até duas casas decimais após a vírgula
pd.options.display.float_format = '{:,.2f}'.format
```

```python
#Agrupando informações por tipo
base.groupby("Tipo")["Curtidas"].mean()
```

```python
#O que os dados indicam é que vídeos e IGTV são estratégias ruins e que não devem ser usadas
```

```python
#Agrupando dados por postagens com pessoas ou não
base.groupby("Pessoas")[["Curtidas", "Comentários"]].mean()
```

```python
#Postagens com pessoas engajam muito mais para essa marca, sendo 3 vezes maior de quando não existem pessoas
```

```python
#Agrupando dados por postagens de campanhas
base.groupby("Campanhas")[["Curtidas","Comentários"]].mean()
```

```python
#Postagens de campanha também tem um engajamento maior!
```

```python
#Agrupando dados de postagens do tipo Carrossel e com filtro de tipo foto
base[base.Tipo == "Foto"].groupby(["Carrossel"])[["Curtidas", "Comentários"]].mean()
```

```python
#A média de curtidas e comentários é maior quando não usamos postagens do tipo carrossel.
```

```python
#Agrupando por pessoas e campanhas
base.groupby(["Pessoas", "Campanhas"])[["Curtidas","Comentários"]].mean()
```

```python
# A média quando temos pessoas na postagem e também são publicações de campanha é de certa de 19,4 mil curtidas, já quando temos apenas pessoas(sem ser publicação de campanha) 
# passamos para quase 10 mil e se não tivermos pessoas a média chega a aproximadamente 5,9 mil curtidas mesmo em campanha.
```

```python
#Agrupando por pessoas, campanhas e tipo
base.groupby(["Pessoas", "Campanhas", "Tipo"])[["Curtidas","Comentários"]].mean()
```

```python
# Vídeos quando feitos em campanha e usando pessoas teve um resultado bom, inclusive próximo ao engajamento de fotos.
# O que poderia ter levado a média para baixo é que só temos vídeos ou com pessoas e de campanha ou sem nenhum dos dois. Não temos vídeos com apenas um dos dois (pessoas ou campanhas).
# Já IGTV, mesmo com pessoas, não teve bons resultados.
```

```python
#Filtrando apenas por postagens do tipo vídeo
base[base.Tipo == "Vídeo"]
```

```python
# Aqui percebemos que a loja tentou postar 4 vídeos mostrando seus produtos (sem nenhuma pessoa) e o resultado foi ruim. Quando o vídeo foi feito com pessoas, o resultado foi muito bom!
```

### ANALISANDO COM BASE NAS TAGS DAS POSTAGENS

```python
#Agrupando por tags
base.groupby("Tags")["Curtidas"].mean()
```

```python
#Algumas postagens contém duas tags, precisamos separá-las para que as médias tragam valores mais reais.
```

```python
#Transformando a coluna Tags em uma lista de tags após quebrar as strings das tags duplas
base.Tags = base.Tags.str.split("/")
base.head()
```

```python
#Separando a coluna Tags em 1 linha para cada elemento da lista com explode
base = base.explode('Tags')
base.head()
```

```python
#Agrupando por tags e ordendando do maior para o menor
base.groupby("Tags")[["Curtidas", "Comentários"]].mean().sort_values("Curtidas", ascending=False)
```

```python
#Postagens de promoções são as que mais engajam
#Além de promoções, datas comemorativas e trends também possuem um bom engajamento
```

```python
#Filtrando valores sem tag
base[base.Tags.isnull()]
```

```python
#Atribuindo valor "Sem tag" para valores NAN na coluna tags
base.loc[base.Tags.isnull(), "Tags"] = "Sem tag"
```

```python
#Agrupando por tags e ordendando do maior para o menor com a nova tag incluída
base.groupby("Tags")[["Curtidas", "Comentários"]].mean().sort_values("Curtidas", ascending=False)
```

```python
#Agrupando pessoas e tag
base.groupby(["Pessoas", "Tags"])[["Curtidas", "Comentários"]].mean().sort_values("Curtidas", ascending=False)
```

```python
#Agrupando campanhas e tag
base.groupby(["Campanhas", "Tags"])[["Curtidas", "Comentários"]].mean().sort_values("Curtidas", ascending=False)
```

### CONCLUSÕES

```python
# Postagens com pessoas engajam muito mais (tanto em foto ou em vídeo) que aquelas que não possuem.
# Postagens em épocas de campanha também possuem um melhor engajamento.
# O carrossel não foi um diferencial para melhorar o engajamento da marca.
# Criar campanhas ajuda muito na divulgação da marca.
# Promoções tiveram um desempenho absurdamente maior que qualquer outra tag. Porém é uma tag que pode trazer custos para a loja, o que deve ser analisado.
# Usar conteúdos que estão em trend também ajudam na divulgação da marca.
# A melhor maneira de mostrar produtos é através de pessoas utilizando-os, e se possível em campanhas de datas comemorativas.
# Para novos produtos a inclusão de pessoas é mais crítica ainda, sendo quase o dobrro de interações.
```
